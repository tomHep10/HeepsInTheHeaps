```python
import matplotlib.pyplot as plt
import numpy as np

# Create an 8x8 grid of zeros (or any constant)
z = np.zeros((8, 8))
z[3, 4] = 0.6

# Plot it with visible grid lines
fig, ax = plt.subplots()

# Add grid lines
ax.set_xticks(np.arange(0, 9, 1))
ax.set_yticks(np.arange(0, 9, 1))
ax.grid(which='both', color='black', linestyle='-', linewidth=1)

# Creates the obstacles
ax.imshow(z, cmap="Greys", extent=[0, 8, 0, 8], vmin=0, vmax=1)

# Turn on minor ticks to make the grid align properly
ax.set_xticks(np.arange(0, 9, 1))
ax.set_yticks(np.arange(0, 9, 1))
ax.grid(color='black', linestyle='-', linewidth=1)

# Hide axes
ax.tick_params(left=False, bottom=False, labelleft=True, labelbottom=True)

plt.title("8x8 Grid")
plt.show()
```
